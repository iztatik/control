## Stability range

```python
import control as ct
import matplotlib.pyplot as plt
```

```python
Kp = float(input('Open-loop gain: '))

num = 1
den = [1, 5, 2, 3, 0]

# Plant
G = ct.tf(num, den)
print(G)
```

```python
# Closed-loop system
sys = ct.feedback(Kp*G, 1)
print(sys)
```

```python
# Poles and zeros map
ct.pzmap(sys)
plt.show()
```

```python
# Root locus
ct.root_locus(G)
plt.show()
```
