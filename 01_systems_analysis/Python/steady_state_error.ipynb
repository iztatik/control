## Steady state error

```python
import control as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plant
num = [38]
den = [1, 5, 6]
G = ct.tf(num, den)
print(G)
```

```python
# Closed-loop system
sys = ct.feedback(G, 1)
print(sys)
```

### System response

```python
# Simulation parameters
SIM_TIME = 4
DT = 0.05
t = np.arange(0, SIM_TIME, DT)
```

```python
# Step response
u = np.ones(t.shape)
_, C1 = ct.forced_response(sys, t, u)

plt.plot(t, u)
plt.plot(t, C1)
plt.grid()
plt.show()
```

```python
# Steady-state error
ess = u[-1] - C1[-1]
print('ess=', ess)
```

```python
# Ramp response
y = t
_, C2 = ct.forced_response(sys, t, y)

plt.plot(t, y)
plt.plot(t, C2)
plt.grid()
plt.show()
```

```python
# Parabola response
p = t**2
_, C3 = ct.forced_response(sys, t, p)

plt.plot(t, p)
plt.plot(t, C3)
plt.grid()
plt.show()
```
